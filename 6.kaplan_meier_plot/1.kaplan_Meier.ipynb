```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter, CoxPHFitter
from sklearn.metrics import roc_curve
from statsmodels.stats.multitest import multipletests  #  p 

def calculate_youden_cutoff(y, scores):
    """
     Youden 
    """
    fpr, tpr, thresholds = roc_curve(y, scores)
    youden_index = tpr - fpr
    best_index = np.argmax(youden_index)
    return thresholds[best_index]


def add_at_risk_table(ax, xticks, low_risk_at_risk, low_risk_events, high_risk_at_risk, high_risk_events):

    table = ax.table(
        cellText=[high_risk_at_risk,high_risk_events,low_risk_at_risk,low_risk_events],
        rowLabels=["High Protein At Risk","High Protein Events","Low Protein At Risk","Low Protein Events"],
        colLabels=[f"{t:.1f}" for t in xticks],
        cellLoc='center',
        loc='bottom',
        bbox=[0.3, -0.4, 0.7, 0.25]
    )
    table.auto_set_font_size(False)
    table.set_fontsize(14)
    table.auto_set_column_width(col=list(range(len(xticks))))
    # 
    for key, cell in table.get_celld().items():
        cell.set_linewidth(0)  #  0


def plot_kaplan_meier(kmf, time, y, low_risk, high_risk, protein, xticks, output_path, cph, pro_data, pro_labels, time_column, covariates):
    """
     Kaplan-Meier  Cox HR95% CI  p 
    """
    plt.figure(figsize=(5, 5))
    ax = plt.gca()

    # 
    kmf.fit(time[low_risk], event_observed=y[low_risk], label=f"{protein} Low")
    kmf.plot(ax=ax, ci_show=True, linestyle='-', color='blue')

    # 
    kmf.fit(time[high_risk], event_observed=y[high_risk], label=f"{protein} High")
    kmf.plot(ax=ax, ci_show=True, linestyle='-', color='red')

    # 
    # plt.title(f"Kaplan-Meier Survival Curve for {protein}")
    plt.xlabel("Follow-up Time (years)",fontsize=16)
    plt.ylabel("Survival Probability",fontsize=16)
    plt.xticks(fontsize = 16)
    plt.yticks(fontsize=16)
    plt.legend(loc="best",fontsize=12)
    plt.grid(alpha=0.3)
    plt.xticks(xticks)

    # 
    # /
    low_risk_at_risk = [sum((time[low_risk] >= t) & (y[low_risk] == 0)) for t in xticks]
    high_risk_at_risk = [sum((time[high_risk] >= t) & (y[high_risk] == 0)) for t in xticks]
    low_risk_events = [sum((time[low_risk] < t) & (y[low_risk] == 1)) for t in xticks]
    high_risk_events = [sum((time[high_risk] < t) & (y[high_risk] == 1)) for t in xticks]

    # 
    # add_at_risk_table(ax, xticks, low_risk_at_risk, low_risk_events, high_risk_at_risk, high_risk_events)

    cox_data = pd.concat([pro_data[[f"{protein}_binary"]], pro_labels[[time_column, "SLE"] + covariates]], axis=1)
    cox_data.columns = [f"{protein}_binary", "Time", "Event"] + covariates
    cph.fit(cox_data, duration_col="Time", event_col="Event")

    hr = cph.hazard_ratios_.loc[f"{protein}_binary"]
    ci_lower = np.exp(cph.confidence_intervals_.loc[f"{protein}_binary", "95% lower-bound"])
    ci_upper = np.exp(cph.confidence_intervals_.loc[f"{protein}_binary", "95% upper-bound"])
    p_value = cph.summary.loc[f"{protein}_binary", "p"]

    _, corrected_p_value, _, _ = multipletests([p_value], method="fdr_bh")
    corrected_p_value = corrected_p_value[0].item()  # 
    corrected_p_value_sci = f"{corrected_p_value:.2e}".replace("e", "10^").replace("^-", "^-")

    text_str = (
        f"HR: {hr:.2f}[{ci_lower:.2f} - {ci_upper:.2f}]\n"
        f"$\\mathit{{P}}$-value: {corrected_p_value_sci}"
    )
    plt.text(
        0.02, 0.02,
        text_str,
        fontsize=10,
        verticalalignment='bottom',
        horizontalalignment='left',
        transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')
    )

    # 
    plt.subplots_adjust(left=0.2, bottom=0.4)
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()


def plot_kaplan_meier_and_cox(pro_data, pro_labels, time_column, covariates, output_dir):
    disease = "SLE"
    kmf = KaplanMeierFitter()
    cph = CoxPHFitter()

    cutoff_dict = {}  #  cutoff 

    for protein in pro_data.columns:
        print(f"Processing protein: {protein}")

        scores = pro_data[protein]
        y = pro_labels[disease]
        time = pro_labels[time_column]
        cutoff = calculate_youden_cutoff(y, scores)
        print(f"Best cutoff for {protein}: {cutoff:.2f}")

        #  cutoff 
        cutoff_dict[protein] = cutoff

        pro_data[f"{protein}_binary"] = (scores >= cutoff).astype(int)

        high_risk = scores >= cutoff
        low_risk = scores < cutoff

        max_time = 15
        # xticks = np.arange(0, max_time + 3, 3)
        xticks = np.arange(0, max_time, 3)

        km_output_path = f"{output_dir}/{protein}_km_curve.pdf"
        plot_kaplan_meier(kmf, time, y, low_risk, high_risk, protein, xticks, km_output_path,
                          cph=CoxPHFitter(), pro_data=pro_data, pro_labels=pro_labels,
                          time_column="FP-SLE", covariates=["SEX", "ETHNICITY"])

    #  cutoff  CSV
    cutoff_df = pd.DataFrame(list(cutoff_dict.items()), columns=['Protein', 'Cutoff'])
    cutoff_df.to_csv(f"{output_dir}/protein_cutoffs.csv", index=False)
    print(f"Cutoff values saved to {output_dir}/protein_cutoffs.csv")

if __name__ == '__main__':
    # 
    pro = pd.read_csv("/Volumes/data_files/UKB_data/processed_data/immune_pro.csv")
    pro = pro.fillna(pro.median(numeric_only=True))
    pro = pro[pro["fp-SLE"] > 0]

    pro.columns = pro.columns.str.upper()
    # 
    pro_data = pro[['TRIM21', 'IL15', 'SCARB2', 'LGALS9', 'PDCD1', 'SOD2', 'MEPE', 'LAG3', 'CXCL16', 'POMC', 'BST2']]
    pro_labels = pro[["SLE", "FP-SLE", "SEX", "ETHNICITY"]]
    time_column = "FP-SLE"
    covariates = ["SEX", "ETHNICITY"]
    output_dir="./km"
    
    #  Kaplan-Meier  Cox 
    plot_kaplan_meier_and_cox(pro_data, pro_labels, time_column, covariates, output_dir)
```
